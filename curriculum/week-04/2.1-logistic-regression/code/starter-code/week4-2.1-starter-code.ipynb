```python
# reserve the first cell, to import all dependencies (python classes) that you will need

import statsmodels.api as sm
import pandas as pd
import pylab as pl
import numpy as np
```

```python
# use function to read in Dataset. 
df = pd.read_csv("http://www.ats.ucla.edu/stat/data/binary.csv")
```

```python
# find data dictionary of df
df
# we want to predict admit 
```

```python
df.columns = list(df.columns)[:3] + ["prestige"]
```

```python
dummy_ranks = pd.get_dummies(df["prestige"], prefix="prestige")
# prestige didnt HAVE to be the dummy to get dropped. 
#You can drop any dummy column and not just prestige
```

```python
cols_to_keep = ["admit", "gre", "gpa"]
data = df[cols_to_keep].join(dummy_ranks.ix[:, "prestige_2":])
```

```python
data["intercept"] = 1.0
train_cols = data.columns[1:]
```

```python
data
```

```python
logit = sm.Logit(data["admit"], data[train_cols])
```

```python
result = logit.fit()
```

```python
result.summary2()
```
